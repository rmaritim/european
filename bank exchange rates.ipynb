```python
import requests
import pandas as pd

# Define the base URL for the API
base_url = "https://api.frankfurter.app"

# Loop over the years from 2013 to 2023
for year in range(2013, 2023):
    try:
        # Send a GET request to the API for the exchange rates on the first day of the year
        response = requests.get(f"{base_url}/{year}-01-01")
        
        # Check if the request was successful (status code 200)
        if response.status_code == 200:
            # Convert the response to JSON
            data = response.json()
            
            # Print the 'rates' data
            print(f"Exchange rates for {year}: {data['rates']}")
        else:
            # If the request was not successful, print an error message
            print(f"Error: Failed to retrieve data for {year}. Status Code: {response.status_code}")
    except Exception as e:
        # If an exception occurs during the request, print the error message
        print(f"Error: {e}")
```

```python
#question 1
import matplotlib.pyplot as plt

# Define the exchange rate data for EUR to USD
exchange_rates = {
    '2013': 1.3194,
    '2014': 1.3791,
    '2015': 1.2141,
    '2016': 1.0887,
    '2017': 1.0541,
    '2018': 1.1993,
    '2019': 1.145,
    '2020': 1.1234,
    '2021': 1.2271,
    '2022': 1.1326
}

# Extract years and exchange rates
years = list(exchange_rates.keys())
rates = list(exchange_rates.values())

# Plotting the line chart
plt.figure(figsize=(10, 6))
plt.plot(years, rates, marker='o', linestyle='-')
plt.title('Exchange Rate of EUR to USD (2013-2022)')
plt.xlabel('Year')
plt.ylabel('Exchange Rate')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()

# Display the line chart
plt.show()
```

```python
#question 3
import numpy as np
import matplotlib.pyplot as plt

# Exchange rates data
exchange_rates_gbp = {
    '2013': 0.8161,
    '2014': 0.8337,
    '2015': 0.7789,
    '2016': 0.73395,
    '2017': 0.85618,
    '2018': 0.88723,
    '2019': 0.89453,
    '2020': 0.8508,
    '2021': 0.89903,
    '2022': 0.84028
}

exchange_rates_jpy = {
    '2013': 113.61,
    '2014': 144.72,
    '2015': 145.23,
    '2016': 131.07,
    '2017': 123.4,
    '2018': 135.01,
    '2019': 125.85,
    '2020': 121.94,
    '2021': 126.49,
    '2022': 130.38
}

# Calculate standard deviations
std_dev_gbp = np.std(list(exchange_rates_gbp.values()))
std_dev_jpy = np.std(list(exchange_rates_jpy.values()))

# Plotting the bar chart
plt.figure(figsize=(8, 6))
plt.bar(['GBP', 'JPY'], [std_dev_gbp, std_dev_jpy], color=['blue', 'green'])
plt.title('Volatility Comparison: GBP vs JPY')
plt.xlabel('Currency')
plt.ylabel('Standard Deviation')
plt.ylim(0, max(std_dev_gbp, std_dev_jpy) * 1.2)
plt.grid(axis='y')
plt.tight_layout()

# Display the bar chart
plt.show()
```

```python
#question 4
import matplotlib.pyplot as plt

# Exchange rates data
exchange_rates_cad = {
    '2013': 1.3137,
    '2014': 1.4671,
    '2015': 1.4063,
    '2016': 1.5116,
    '2017': 1.4188,
    '2018': 1.5039,
    '2019': 1.5605,
    '2020': 1.4598,
    '2021': 1.5633,
    '2022': 1.4393
}

exchange_rates_usd = {
    '2013': 1.3194,
    '2014': 1.3791,
    '2015': 1.2141,
    '2016': 1.0887,
    '2017': 1.0541,
    '2018': 1.1993,
    '2019': 1.145,
    '2020': 1.1234,
    '2021': 1.2271,
    '2022': 1.1326
}

# Convert dictionaries to lists for easier plotting
years = list(exchange_rates_cad.keys())
exchange_rates_cad_values = list(exchange_rates_cad.values())
exchange_rates_usd_values = list(exchange_rates_usd.values())

# Plotting the scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(exchange_rates_cad_values, exchange_rates_usd_values, color='blue')
plt.title('Exchange Rates: CAD vs USD')
plt.xlabel('CAD')
plt.ylabel('USD')
plt.grid(True)
plt.tight_layout()

# Adding year labels to the data points
for i, year in enumerate(years):
    plt.annotate(year, (exchange_rates_cad_values[i], exchange_rates_usd_values[i]), textcoords="offset points", xytext=(10,0), ha='center')

# Display the scatter plot
plt.show()
```
